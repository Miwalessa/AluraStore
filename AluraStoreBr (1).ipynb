### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
faturamento_loja1 = loja['Preço'].sum()
faturamento_loja2 = loja2['Preço'].sum()
faturamento_loja3 = loja3['Preço'].sum()
faturamento_loja4 = loja4['Preço'].sum()

print(f"Faturamento da Loja 1 é: {faturamento_loja1}")
print(f"Faturamento da Loja 2 é: {faturamento_loja2}")
print(f"Faturamento da Loja 3 é: {faturamento_loja3}")
print(f"Faturamento da Loja 4 é: {faturamento_loja4}")
```

```python
import matplotlib.pyplot as plt


labels = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
sizes = [faturamento_loja1, faturamento_loja2, faturamento_loja3, faturamento_loja4]
colors = ['gold', 'lightblue', 'lightcoral', 'lightgreen']
explode = (0.1, 0, 0, 0)


plt.figure(figsize=(7, 7))
plt.pie(sizes, labels=labels, colors=colors, explode=explode, autopct='%1.1f%%', shadow=True, startangle=140)

plt.title('Faturamento das Lojas')
plt.show()
```

# 2. Vendas por Categoria


```python
total_categoria_loja1 = loja.groupby("Categoria do Produto")["Preço"].sum().reset_index()
total_categoria_loja2 = loja2.groupby("Categoria do Produto")["Preço"].sum().reset_index()
total_categoria_loja3 = loja3.groupby("Categoria do Produto")["Preço"].sum().reset_index()
total_categoria_loja4 = loja4.groupby("Categoria do Produto")["Preço"].sum().reset_index()

print(f"Loja 1:\n{total_categoria_loja1}\n")
print(f"Loja 2:\n{total_categoria_loja2}\n")
print(f"Loja 3:\n{total_categoria_loja3}\n")
print(f"Loja 4:\n{total_categoria_loja4}\n")
```

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(12, 6))


plt.plot(total_categoria_loja1["Categoria do Produto"], total_categoria_loja1["Preço"], marker='o', linestyle='-', label="Loja 1", color='blue')
plt.plot(total_categoria_loja2["Categoria do Produto"], total_categoria_loja2["Preço"], marker='s', linestyle='--', label="Loja 2", color='red')
plt.plot(total_categoria_loja3["Categoria do Produto"], total_categoria_loja3["Preço"], marker='^', linestyle='-.', label="Loja 3", color='green')
plt.plot(total_categoria_loja4["Categoria do Produto"], total_categoria_loja4["Preço"], marker='d', linestyle=':', label="Loja 4", color='orange')

plt.xticks(rotation=45)
plt.ylabel("Faturamento")
plt.xlabel("Categoria do Produto")
plt.title("Evolução do Faturamento por Categoria nas Lojas")
plt.legend()
plt.grid(True)

plt.show()
```

# 3. Média de Avaliação das Lojas

```python
media_avaliacao_loja1 = loja["Avaliação da compra"].mean()
media_avaliacao_loja2 = loja2["Avaliação da compra"].mean()
media_avaliacao_loja3 = loja3["Avaliação da compra"].mean()
media_avaliacao_loja4 = loja4["Avaliação da compra"].mean()

print(f"Média de avalição loja 1: {media_avaliacao_loja1:.2f}")
print(f"Média de avalição loja 2: {media_avaliacao_loja2:.2f}")
print(f"Média de avalição loja 3: {media_avaliacao_loja3:.2f}")
print(f"Média de avalição loja 4: {media_avaliacao_loja4:.2f}")
```

```python
import matplotlib.pyplot as plt


lojas = ["Loja 1", "Loja 2", "Loja 3", "Loja 4"]
medias = [media_avaliacao_loja1, media_avaliacao_loja2, media_avaliacao_loja3, media_avaliacao_loja4]


plt.figure(figsize=(8, 5))
bars = plt.bar(lojas, medias, color=['blue', 'red', 'green', 'orange'])


for bar, media in zip(bars, medias):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(), f"{media:.2f}", ha='center', va='bottom', fontsize=12, fontweight='bold', color='black')


plt.xlabel("Lojas")
plt.ylabel("Média de Avaliação")
plt.title("Comparação das Avaliações por Loja")
plt.ylim(0, 5)

plt.show()
```

# 4. Produtos Mais e Menos Vendidos

```python
quantidade_produto_loja1 = loja["Produto"].value_counts().reset_index()
quantidade_produto_loja2 = loja2["Produto"].value_counts().reset_index()
quantidade_produto_loja3 = loja3["Produto"].value_counts().reset_index()
quantidade_produto_loja4 = loja4["Produto"].value_counts().reset_index()


print(f"Os 10 produtos mais vendidos da loja1:\n {quantidade_produto_loja1.head(10)}\n E os 10 produtos menos vendidos:\n {quantidade_produto_loja1.tail(10)}")
print(f"Total de vendas de cada produto loja2:\n {quantidade_produto_loja2.head(10)}\n E os 10 produtos menos vendidos:\n {quantidade_produto_loja2.tail(10)}\n")
print(f"Total de vendas de cada produto loja3:\n {quantidade_produto_loja3.head(10)}\n E os 10 produtos menos vendidos:\n {quantidade_produto_loja3.tail(10)}\n")
print(f"Total de vendas de cada produto loja4:\n {quantidade_produto_loja4.head(10)}\n E os 10 produtos menos vendidos:\n {quantidade_produto_loja4.tail(10)}\n")
```

```python
import matplotlib.pyplot as plt


quantidade_produto_loja1.columns = ["Produto", "Quantidade"]
quantidade_produto_loja2.columns = ["Produto", "Quantidade"]
quantidade_produto_loja3.columns = ["Produto", "Quantidade"]
quantidade_produto_loja4.columns = ["Produto", "Quantidade"]


produtos_mais_vendidos_loja1 = quantidade_produto_loja1.head(10)
produtos_menos_vendidos_loja1 = quantidade_produto_loja1.tail(10)

produtos_mais_vendidos_loja2 = quantidade_produto_loja2.head(10)
produtos_menos_vendidos_loja2 = quantidade_produto_loja2.tail(10)

produtos_mais_vendidos_loja3 = quantidade_produto_loja3.head(10)
produtos_menos_vendidos_loja3 = quantidade_produto_loja3.tail(10)

produtos_mais_vendidos_loja4 = quantidade_produto_loja4.head(10)
produtos_menos_vendidos_loja4 = quantidade_produto_loja4.tail(10)


fig, axes = plt.subplots(4, 2, figsize=(14, 12))


axes[0, 0].barh(produtos_mais_vendidos_loja1["Produto"], produtos_mais_vendidos_loja1["Quantidade"], color='blue')
axes[0, 0].set_title("Top 10 - Loja 1")

axes[1, 0].barh(produtos_mais_vendidos_loja2["Produto"], produtos_mais_vendidos_loja2["Quantidade"], color='red')
axes[1, 0].set_title("Top 10 - Loja 2")

axes[2, 0].barh(produtos_mais_vendidos_loja3["Produto"], produtos_mais_vendidos_loja3["Quantidade"], color='green')
axes[2, 0].set_title("Top 10 - Loja 3")

axes[3, 0].barh(produtos_mais_vendidos_loja4["Produto"], produtos_mais_vendidos_loja4["Quantidade"], color='orange')
axes[3, 0].set_title("Top 10 - Loja 4")


axes[0, 1].barh(produtos_menos_vendidos_loja1["Produto"], produtos_menos_vendidos_loja1["Quantidade"], color='blue')
axes[0, 1].set_title("Menos Vendidos - Loja 1")

axes[1, 1].barh(produtos_menos_vendidos_loja2["Produto"], produtos_menos_vendidos_loja2["Quantidade"], color='red')
axes[1, 1].set_title("Menos Vendidos - Loja 2")

axes[2, 1].barh(produtos_menos_vendidos_loja3["Produto"], produtos_menos_vendidos_loja3["Quantidade"], color='green')
axes[2, 1].set_title("Menos Vendidos - Loja 3")

axes[3, 1].barh(produtos_menos_vendidos_loja4["Produto"], produtos_menos_vendidos_loja4["Quantidade"], color='orange')
axes[3, 1].set_title("Menos Vendidos - Loja 4")


for ax in axes.flat:
    ax.set_xlabel("Quantidade Vendida")
    ax.set_ylabel("Produto")
    ax.invert_yaxis()

plt.tight_layout()
plt.show()
```

# 5. Frete Médio por Loja


```python
media_frete_loja1 = loja['Frete'].mean()
media_frete_loja2 = loja2['Frete'].mean()
media_frete_loja3 = loja3['Frete'].mean()
media_frete_loja4 = loja4['Frete'].mean()

print(f"Média frete loja1: {media_frete_loja1:.2f}")
print(f"Média frete loja2: {media_frete_loja2:.2f}")
print(f"Média frete loja3: {media_frete_loja3:.2f}")
print(f"Média frete loja4: {media_frete_loja4:.2f}")
```

```python
import matplotlib.pyplot as plt


lojas = ["Loja 1", "Loja 2", "Loja 3", "Loja 4"]
medias_frete = [media_frete_loja1, media_frete_loja2, media_frete_loja3, media_frete_loja4]


plt.figure(figsize=(8, 5))
bars = plt.bar(lojas, medias_frete, color=['blue', 'red', 'green', 'orange'])


for bar, media in zip(bars, medias_frete):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(), f"{media:.2f}", ha='center', va='bottom', fontsize=12, fontweight='bold', color='black')


plt.xlabel("Lojas")
plt.ylabel("Média do Frete")
plt.title("Comparação da Média do Frete por Loja")
plt.ylim(0, max(medias_frete) * 1.2)

plt.show()
```
